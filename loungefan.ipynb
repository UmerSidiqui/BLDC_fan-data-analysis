```python
import pandas as pd
import numpy as np
import matplotlib as pb
import matplotlib.pyplot as plt
```

```python
data_lounge1=pd.read_csv("Lounge_Fan_1.csv")
data_lounge2=pd.read_csv("Lounge_Fan_2.csv")
data_lounge3=pd.read_csv("Lounge_Fan_3.csv")
```

```python
data_lounge1=pd.DataFrame(data_lounge1)
data_lounge2=pd.DataFrame(data_lounge2)
data_lounge3=pd.DataFrame(data_lounge3)
data_lounge2=data_lounge2[data_lounge2['temperature']<50]
```

```python
# hourly data with max features

import pandas as pd

# Assuming df is your existing DataFrame

# Step 1: Convert the 'datetime' column to datetime objects
data_lounge2['datetime'] = pd.to_datetime(data_lounge2['datetime'], format='%d/%m/%Y %H:%M')

# Step 2: Extract the hour part from the 'datetime' column
data_lounge2['hour'] = data_lounge2['datetime'].dt.floor('H')  # Round down to the nearest hour

# Step 3: Group by the hour and calculate the mean for each group
hourly_mode_df = data_lounge2.groupby('hour').max().reset_index()

# Optionally, you may want to drop the 'S.No' column from the resulting DataFrame
hourly_mode_df = hourly_mode_df.drop(columns=['S.No.'])  # Drop the 'S.No' column if you do not need it

# Output the resulting DataFrame
hourly_mode_df
```

```python
import plotly.express as px

# Assuming your data is stored in a DataFrame called df
# Replace 'temperature' and 'speed' with the actual column names from your DataFrame
fig = px.line(hourly_mode_df, x='hour', y='temperature', title='time vs temperature',
                 labels={'hour': 'time', 'temperature': 'temperature'})

fig.show()
```

```python
# daily data with max features

import pandas as pd

# Assuming df is your existing DataFrame

# Step 1: Convert the 'datetime' column to datetime objects
data_lounge2['datetime'] = pd.to_datetime(data_lounge2['datetime'], format='%d/%m/%Y %H:%M')

# Step 2: Extract the hour part from the 'datetime' column
data_lounge2['day'] = data_lounge2['datetime'].dt.floor('D')  # Round down to the nearest hour

# Step 3: Group by the hour and calculate the mean for each group
daily_mode_df = data_lounge2.groupby('day').max().reset_index()

# Optionally, you may want to drop the 'S.No' column from the resulting DataFrame
daily_mode_df = daily_mode_df.drop(columns=['S.No.'])  # Drop the 'S.No' column if you do not need it

# Output the resulting DataFrame
daily_mode_df 
```

```python
import plotly.express as px

# Assuming your data is stored in a DataFrame called df
# Replace 'temperature' and 'speed' with the actual column names from your DataFrame
fig = px.line(daily_mode_df, x='day', y='temperature', title='time vs temperature',
                 labels={'day': 'time', 'temperature': 'temperature'})

fig.show()
```

```python
# prophet model for temp
```

```python
ds_df_t= pd.DataFrame(daily_mode_df[['day','temperature']])

ds_df_t = ds_df_t.rename(columns={'day': 'ds','temperature':'y'})


# prophet model
from prophet import Prophet
model_t = Prophet()
model_t.fit(ds_df_t)

future_dates_temp = model_t.make_future_dataframe(periods=30, freq='D') 


forecast_temp = model_t.predict(future_dates_temp)


fig = model_t.plot(forecast_temp)
```

```python
ds_df_h= pd.DataFrame(daily_mode_df[['day','humidity']])

ds_df_h = ds_df_h.rename(columns={'day': 'ds','humidity':'y'})


# prophet model
from prophet import Prophet
model_h = Prophet()
model_h.fit(ds_df_h)

future_dates_humid = model_h.make_future_dataframe(periods=30, freq='D') 


forecast_humid = model_h.predict(future_dates_humid)


fig = model_h.plot(forecast_humid)
```

```python
error_series=forecast_temp.iloc[:315]['temperature']-daily_mode_df['temperature']
# Mean Absolute Error (MAE)
mae = error_series.abs().mean()

# Mean Squared Error (MSE)
mse = (error_series ** 2).mean()

# Root Mean Squared Error (RMSE)
rmse = np.sqrt(mse)

# Output the results
print(f"Mean Absolute Error (MAE): {mae}")
print(f"Mean Squared Error (MSE): {mse}")
print(f"Root Mean Squared Error (RMSE): {rmse}")
```

```python
error_series=forecast_humid.iloc[:315]['humidity']-daily_mode_df['humidity']
# Mean Absolute Error (MAE)
mae = error_series.abs().mean()

# Mean Squared Error (MSE)
mse = (error_series ** 2).mean()

# Root Mean Squared Error (RMSE)
rmse = np.sqrt(mse)

# Output the results
print(f"Mean Absolute Error (MAE): {mae}")
print(f"Mean Squared Error (MSE): {mse}")
print(f"Root Mean Squared Error (RMSE): {rmse}")
```
